```python
#文本就应该有一个文本读取的样子，不能用series来整理未分类数据,这里学会了如何将乱七八糟的txt文件转化成dataframe，主要是key-value配对
f=open('df1')
df1={'date':[],'open':[],'high':[],'low':[],'close':[]}
for each_line in f:
    each_line=f.readline()
    line=each_line.strip().split('\t')
    for i in range(0,len(line)):
        df1['date'].append(line[0])
        df1['open'].append(line[1])
        df1['high'].append(line[2])
        df1['low'].append(line[3])
        df1['close'].append(line[4])
df1=pd.DataFrame(df1)
df1.set_index(["date"], inplace=True)

for i in df1.columns:
    df1[i]=pd.to_numeric(df1[i])
print(df1)
print(type(df1.columns))
print(type(df1['open'][0]))
```

```python
f2=open('df2')
df2={'date':[],'open':[],'high':[],'low':[],'close':[]}
for each_line in f2:
    each_line=f2.readline()
    line=each_line.strip().split('\t')
    for i in range(0,len(line)):
        df2['date'].append(line[0])
        df2['open'].append(line[1])
        df2['high'].append(line[2])
        df2['low'].append(line[3])
        df2['close'].append(line[4])
df2=pd.DataFrame(df2)
df2.set_index(["date"], inplace=True)
for i in df2.columns:
    df2[i]=pd.to_numeric(df2[i])
print(df2)
```

```python
#add color map
df1.plot.scatter(x='high',y='close',c='open',cmap='coolwarm') #这里的c是以open作为color map， 红色的是close，x是high

'''
1. 什么是色彩条？
色彩条其实是另一条y轴，只不过不是刻度，是颜色，从冷色到暖色，那你的线越高颜色就越暖。
2. 为什么要用色彩条？
在图像处理，尤其是医学图像处理的过程中，我们经常会遇到将灰度图映射成彩色图的情形，
如将灰度图根据灰度的高低映射成彩虹色图。
这个过程我们通常将之称为伪彩映射，伪彩映射的关键在于找到合适的彩色映射表，即colormap，也称color bar。

'''
```

```python
#对于散点图scatter，不仅可以设置点的颜色，也可以设置点的相对大小,s代表size
'''我觉得这个很有意思，很直观，
scatter不仅比较了两个数据之间的关系，还可以显示出某个数据的相对大小
'''
df1.plot.scatter(x='high',y='close',s=df1['open']*50)
```
