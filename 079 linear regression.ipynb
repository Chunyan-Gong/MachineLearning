```python
X=f[['Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y=f['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)  #this is unpacking

'''
random_state ensures a specific set of random splits
'''
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#create a LR object
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```
